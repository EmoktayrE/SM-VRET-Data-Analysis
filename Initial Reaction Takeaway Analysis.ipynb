```python
df = pd.read_csv('Session Recording Coding Categories.csv')
df
```

```python
catList = list(df['Categories'].unique())
catList
```

```python
df[['Codes', 'Categories']].value_counts().head(6)
```

```python
catConfusion = df['Categories'][df['Categories'] == 'Confusion'].count()
catPositive = df['Categories'][df['Categories'] == 'Positive'].count()
catNegative = df['Categories'][df['Categories'] == 'Negative'].count()
catBoredom = df['Categories'][df['Categories'] == 'Boredom'].count()

```

```python
catCountdata = {'Categories': catList, 'Code Count': [catConfusion, catPositive, catNegative, catBoredom]}

catCountdf = pd.DataFrame(data = catCountdata)

catCountdf = catCountdf.sort_values(by = ['Code Count'], ascending = False, ignore_index = True)

print(catCountdf.to_latex())
```

It can be observed that from the session recordings, most of the initial reactions were positive. However, the negative implications were close to the number of the positive ones. The confusion category exists since there were a number of instances where evaluators were confused about particular attributes within the application. And boredom is a category since at instances, tiresome comments were made for selected elements. 

If confusion and boredom are considered negative however, negative comments get ahead of the positives by 33 to 21.

```python
#confusion table

conVC = df[df['Categories'] == 'Confusion'].value_counts()
Cdf = pd.DataFrame(conVC)
Cdf = Cdf.reset_index()
Cdf = Cdf.rename(columns = {0:'Code Count'})
Cdf
```

```python
# positive table

posVC = df[df['Categories'] == 'Positive'].value_counts()
Pdf = pd.DataFrame(posVC)
Pdf = Pdf.reset_index()
Pdf = Pdf.rename(columns = {0:'Code Count'})
Pdf
```

```python
# negative table

negVC = df[df['Categories'] == 'Negative'].value_counts()
Ndf = pd.DataFrame(negVC)
Ndf = Ndf.reset_index()
Ndf = Ndf.rename(columns = {0:'Code Count'})
Ndf
```

```python
# boredom table

borVC = df[df['Categories'] == 'Boredom'].value_counts()
Bdf = pd.DataFrame(borVC)
Bdf = Bdf.reset_index()
Bdf = Bdf.rename(columns = {0:'Code Count'})
Bdf
```

```python
fullDF = pd.concat([Cdf, Pdf, Ndf, Bdf])
fullDF
```

```python
print(fullDF.to_latex())
```
