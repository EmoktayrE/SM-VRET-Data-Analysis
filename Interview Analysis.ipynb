```python
df = pd.read_csv('Interviews/Interview Coding Categories.csv')
pd.set_option('display.max_rows', 100)
df
```

```python
catList = list(df['Categories'].unique())
```

```python
# code count with categories

df[['Codes', 'Categories']].value_counts()
```

```python
catPositive = df['Categories'][df['Categories'] == 'Positive Implications'].count()
catNegative = df['Categories'][df['Categories'] == 'Negative Implications'].count()
catExtension = df['Categories'][df['Categories'] == 'Ideas for Extension'].count()
catNegInteraction = df['Categories'][df['Categories'] == 'Negative Children Interaction'].count()
catAdvice = df['Categories'][df['Categories'] == 'Advice for Improving Evaluation'].count()
catFuture = df['Categories'][df['Categories'] == 'Future Work Questions'].count()
```

### Category Count DataFrame

```python
catCountdata = {'Categories': catList, 'Code Count': [catPositive, catNegative, catExtension, 
                                                                         catNegInteraction, catAdvice, catFuture]}

catCountdf = pd.DataFrame(data = catCountdata).reset_index(drop = True)
catCountdf



```

```python
#positive implications table

# positive table
posVC = df[df['Categories'] == 'Positive Implications'].value_counts()
Pdf = pd.DataFrame(posVC)
Pdf = Pdf.reset_index()
Pdf = Pdf.rename(columns = {0:'Code Count'})
Pdf
```

```python
# negative implications table

negVC = df[df['Categories'] == 'Negative Implications'].value_counts()
Ndf = pd.DataFrame(negVC)
Ndf = Ndf.reset_index()
Ndf = Ndf.rename(columns = {0:'Code Count'})
Ndf
```

```python
# ideas fir extension table

ideaVC = df[df['Categories'] == 'Ideas for Extension'].value_counts()
Idf = pd.DataFrame(ideaVC)
Idf = Idf.reset_index()
Idf = Idf.rename(columns = {0:'Code Count'})
Idf
```

```python
# negative children interaction table

neginterVC = df[df['Categories'] == 'Negative Children Interaction'].value_counts()
NIdf = pd.DataFrame(neginterVC)
NIdf = NIdf.reset_index()
NIdf = NIdf.rename(columns = {0:'Code Count'})
NIdf
```

```python
# Advice for Improving Evaluation

impVC = df[df['Categories'] == 'Advice for Improving Evaluation'].value_counts()
Adf = pd.DataFrame(impVC)
Adf = Adf.reset_index()
Adf = Adf.rename(columns = {0:'Code Count'})
Adf
```

```python
# Future Work Questions

futureVC = df[df['Categories'] == 'Future Work Questions'].value_counts()
FWdf = pd.DataFrame(futureVC)
FWdf = FWdf.reset_index()
FWdf = FWdf.rename(columns = {0:'Code Count'})
FWdf
```

```python
fullDF = pd.concat([Pdf, Ndf, Idf, NIdf, Adf, FWdf])
fullDF
```
